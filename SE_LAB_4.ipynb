```python
import numpy as np
import matplotlib.pyplot as plt


def quadratic(x):
    a = -1
    return a * x**2 + b * x + c


stages = ["Requirements", "Design", "Implementation", "Verification", "Maintenance"]


y_positions = np.linspace(len(stages) - 1, 0, len(stages))

x_positions = quadratic(y_positions)


y_curve = np.linspace(-2.5, 2.5, 500)

x_curve = quadratic(y_curve)


plt.figure(figsize=(10, 6))
plt.plot(x_curve, y_curve, color="blue", linestyle="--", label="Quadratic Curve")


for i in range(len(stages)):
    if i > 0:

        plt.plot([x_positions[i-1], x_positions[i]], [y_positions[i-1], y_positions[i]], color="blue", linewidth=2)

    plt.scatter(x_positions[i], y_positions[i], color="red", zorder=5)


    plt.text(x_positions[i] + 0.1, y_positions[i], stages[i], va="center", fontsize=10)


plt.ylabel('Stages')
plt.xlabel("Quadratic Model Output")
plt.title("Waterfall Model Representation with Quadratic Function")
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()


```
